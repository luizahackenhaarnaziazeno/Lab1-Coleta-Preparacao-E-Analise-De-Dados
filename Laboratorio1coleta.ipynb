# **HTML da página inicial escolhida**

*Utilizando a biblioteca requests sem tratamento de erros.*

```python
import requests
from bs4 import BeautifulSoup

pagina = "https://pt.wikipedia.org/wiki/Ada_Lovelace"

# Requisição sem tratamento de erros
requisicao = requests.get(pagina, headers={'User-Agent': 'Lu e Gabi'})
print(requisicao)

# Criando objeto BeautifulSoup
soup = BeautifulSoup(requisicao.text, "html.parser")
```

```python
import os

#Html da página incial
print(requisicao.text)

# Criação da pasta Html
output_dir = "Htmls"
os.makedirs(output_dir, exist_ok=True)

# Salva o HTML da página inicial em um arquivo .html
with open(os.path.join(output_dir, "Htmlpaginainicial.html"), "w", encoding="utf-8") as arquivo:
  arquivo.write(requisicao.text)
```

#**Scraping dos HTMLs baixados**

***Título da página***

```python
titulo_pagina = soup.head.title.string
print("Título da página: ",titulo_pagina, "\n")
```

***Primeiro parágrafo do artigo***

```python
primeiro_paragrafo = soup.find('p')
print("Primeiro parágrafo: ", primeiro_paragrafo.text, "\n")
```

***Lista de links para cada imagem do artigo***

```python
links_imagens = []
for img in soup.select("img"):
  if img.has_attr("src"):
    src = img["src"]
    links_imagens.append(src)

print("Número de imagens encontradas: ", len(links_imagens))
print("Links das imagens encontradas: ", links_imagens)
```

***Lista de links internos***
> (apenas /wiki/)



*Utilizando seletores complexos -> a[href^='/wiki/']*

```python
links_internos = []
for a in soup.select("a[href^='/wiki/']"):
  link = "https://pt.wikipedia.org" + a["href"]
  links_internos.append(link)

print("Número de links internos: ", len(links_internos))
print(links_internos)
```





# HTMLs de todos os links internos encontrados nesta página
> OBS: apenas links que começam com a tag /wiki/

> OBS: Executar lista de links internos antes desse comando





```python
output_dir_htmls = "Htmls"
os.makedirs(output_dir_htmls, exist_ok=True)

saved_files_count = 0

for i, link_interno in enumerate(links_internos):
    try:
        print(f"\n--- Coletando HTML do Link Interno {i+1}: {link_interno} ---")
        requisicao_link_interno = requests.get(link_interno, headers={'User-Agent': 'Lu e Gabi'})
        requisicao_link_interno.raise_for_status() # HTTPError para respostas HTTP inválidas (4xx ou 5xx)

        filename_base = link_interno.replace('https://pt.wikipedia.org/wiki/', '').replace('/', '_').replace(':', '_').replace('?', '_').replace('=', '_')
        if not filename_base:
            filename_base = f"link_interno_{i+1}"
        filename = os.path.join(output_dir_htmls, f"{filename_base}.html")

        with open(filename, "w", encoding="utf-8") as f:
            f.write(requisicao_link_interno.text)
        print(f"HTML salvo em: {filename}")
        saved_files_count += 1

    except requests.exceptions.RequestException as e:
        print(f"Erro ao acessar ou salvar o link {link_interno}: {e}")

print(f"\nTotal de arquivos HTML salvos: {saved_files_count}")
print(f"Total de links internos processados: {len(links_internos)}")
```

#**Exportação para CSV**

Lista de links para cada imagem do artigo (corrigido)

```python
#Acessando os links completos das imagens
links_corrigidos_imagens = []
for link in links_imagens:
  if link.startswith("//"):
    link = "https:" + link
  elif link.startswith("/"):
    link = "https://pt.wikipedia.org" + link
  links_corrigidos_imagens.append(link)

print("Número de imagens encontradas: ", len(links_corrigidos_imagens))

print("Links corrigidos:")
print(links_corrigidos_imagens)
```

```python
import csv
import os

# Criação da pasta csv
output_dir = "csv"
os.makedirs(output_dir, exist_ok=True)

with open(os.path.join(output_dir, "links_imagens.csv"), "w") as arquivo:
  linha = csv.writer(arquivo)
  linha.writerow(["imagem_url"])
  for img in links_corrigidos_imagens:
    linha.writerow([img])
```

Outro CSV contendo os dados de cada página descritos acima

```python
import unicodedata #Biblioteca usada para deixar o csv mais bonito,fazendo a separação das letras dos acentos para podermos apagar os acentos depois.

# Função auxiliar para remover acentuação
def remover_acentos(texto):
    if not isinstance(texto, str):
        return texto
    return unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')

# Limpa o título e o parágrafo
titulo_limpo = remover_acentos(titulo_pagina)
paragrafo_limpo = remover_acentos(primeiro_paragrafo.text if primeiro_paragrafo else "")


output_file_path = os.path.join(output_dir, "dados_pagina_ada_lovelace.csv")


with open(output_file_path, "w", newline='', encoding='utf-8-sig') as arquivo_csv:
    writer = csv.writer(arquivo_csv, delimiter=';')

    # Escreve o cabeçalho
    writer.writerow([
        "titulo_pagina",
        "primeiro_paragrafo",
        "links_internos",
        "links_imagens"
    ])

    # Descobre qual lista de links é a maior para saber quantas linhas criar
    qtd_links_internos = len(links_internos)
    qtd_links_imagens = len(links_corrigidos_imagens)
    max_linhas = max(qtd_links_internos, qtd_links_imagens)

    # Se a página não tiver absolutamente nenhum link, garantimos 1 linha
    # só para salvar o título e o parágrafo
    if max_linhas == 0:
        max_linhas = 1

    # Cria uma nova linha no CSV para cada link encontrado
    for i in range(max_linhas):
        # O título e o parágrafo só aparecem na PRIMEIRA linha (quando i é 0)
        # Nas linhas seguintes, essas colunas ficarão em branco ("")
        col_titulo = titulo_limpo if i == 0 else ""
        col_paragrafo = paragrafo_limpo if i == 0 else ""

        # Pega o link atual se a lista ainda tiver itens, senão deixa em branco
        col_link_int = remover_acentos(links_internos[i]) if i < qtd_links_internos else ""
        col_link_img = remover_acentos(links_corrigidos_imagens[i]) if i < qtd_links_imagens else ""

        # Escreve a linha no arquivo
        writer.writerow([
            col_titulo,
            col_paragrafo,
            col_link_int,
            col_link_img
        ])
```

Inclua uma coluna extra com o timestamp do momento da coleta
*(datetime.now())*

```python
from datetime import datetime, timedelta, timezone

# Cria um fuso horário personalizado de UTC-3 (Horário de Brasília)
fuso_br = timezone(timedelta(hours=-3))

# Captura o momento da coleta com fuso correto
timestamp_atual = datetime.now(fuso_br).strftime("%Y-%m-%d %H:%M:%S")


output_file_path = os.path.join(output_dir, "dados_pagina_ada_lovelace_new.csv")

with open(output_file_path, "w", newline='', encoding='utf-8-sig') as arquivo_csv:
    writer = csv.writer(arquivo_csv, delimiter=';')

    # Escreve o cabeçalho com a nova coluna
    writer.writerow([
        "data_coleta",
        "titulo_pagina",
        "primeiro_paragrafo",
        "links_internos",
        "links_imagens"
    ])

    # Cria uma nova linha no CSV para cada link encontrado
    for i in range(max_linhas):
        # A data, o título e o parágrafo só aparecem na PRIMEIRA linha
        col_timestamp = timestamp_atual if i == 0 else ""
        col_titulo = titulo_limpo if i == 0 else ""
        col_paragrafo = paragrafo_limpo if i == 0 else ""

        # Pega o link atual se a lista ainda tiver itens
        col_link_int = remover_acentos(links_internos[i]) if i < qtd_links_internos else ""
        col_link_img = remover_acentos(links_corrigidos_imagens[i]) if i < qtd_links_imagens else ""

        # Escreve a linha no arquivo
        writer.writerow([
            col_timestamp,
            col_titulo,
            col_paragrafo,
            col_link_int,
            col_link_img
        ])
```

***Baixando as pastas criadas***

```python
import shutil
from google.colab import files

# Define o nome do arquivo zip
zip_filename = 'CPALab1Tarefa1.zip'

# Cria um diretório temporário para agrupar as pastas
temp_dir = 'temp_download_content'
os.makedirs(temp_dir, exist_ok=True)

# Copia as pastas 'Htmls' e 'csv' para o diretório temporário
# dirs_exist_ok=True permite a cópia mesmo que o destino já exista
if os.path.exists('Htmls'):
    shutil.copytree('Htmls', os.path.join(temp_dir, 'Htmls'), dirs_exist_ok=True)
if os.path.exists('csv'):
    shutil.copytree('csv', os.path.join(temp_dir, 'csv'), dirs_exist_ok=True)

# Cria o arquivo zip a partir do diretório temporário
# O primeiro argumento é o caminho base para o arquivo zip (sem a extensão)
# O segundo é o formato ('zip')
# O terceiro é o diretório raiz a ser compactado
shutil.make_archive(os.path.splitext(zip_filename)[0], 'zip', temp_dir)

# Remove o diretório temporário
shutil.rmtree(temp_dir)

# Agora o arquivo zip deve existir e pode ser baixado
files.download(zip_filename)
```
